```python
from transformers import MvpTokenizer, MvpForConditionalGeneration
import gradio
tokenizer = MvpTokenizer.from_pretrained("RUCAIBox/mvp-story")
model = MvpForConditionalGeneration.from_pretrained("RUCAIBox/mvp-story")
```

```python
def story_generation(input_prompt):
  inputs = tokenizer(input_prompt ,return_tensors="pt",padding=True)
  generated_ids = model.generate(**inputs, max_length=60)
  return tokenizer.batch_decode(generated_ids, skip_special_tokens=True)
```

```python
demo = gradio.Interface(
    fn=story_generation,
    inputs=["text"],
    outputs=["text"],
    )
demo.launch()
```
